```r
if(! require(pacman) )
    install.packages("pacman")
```

```r
pacman::p_load(tidyverse, visdat, esquisse, inspectdf, plotly, janitor)
```

```r
ebola_data <- read.csv("ebola_sierra_leone.csv")

```

```r
ebola_data$age # mean age
```

```r
mean(ebola_data$age) # mean age
```

```r
mean(ebola_data$age, na.rm = TRUE) # to remove NA values
```

```r
median(ebola_data$age, na.rm = TRUE) 
```

```r
sd(ebola_data$age, na.rm=TRUE)# standard deviation
```

```r
min(ebola_data$age, na.rm=TRUE) # minimum age
```

```r
age_vector <- ebola_data$age
```

```r
max(age_vector, na.rm = TRUE) # maximum age
```

```r
summary(age_vector) # summary statistics
```

```r
length(age_vector) # number of observations
```

```r
sum(age_vector, na.rm = TRUE) # sum of ages
```

```r
hist(age_vector) # histogram of ages
```

```r
boxplot(age_vector) # boxplot of ages
```

```r
esquisse::esquisser(ebola_data) # interactive plotting with esquisse
```

```r
pacman::p_load(ggplot2)
```

```r
ggplot(ebola_data) +
  aes(x = age, y = sex, fill = sex) +
  geom_bar(stat = "summary", fun = "sum") +
  scale_fill_hue(direction = 1) +
  theme_minimal()
```

```r
district_vector <- ebola_data$district   # frequency table of districts
```

```r
table(district_vector) # frequency table of districts
```

```r
janitor::tabyl(district_vector) # frequency table with janitor
```

```r
janitor::tabyl(ebola_data, district) # frequency table with janitor directly on data frame
```

```r
janitor::tabyl(ebola_data, district, sex) # cross-tabulation of district and sex: which means how many males and females in each district
```

```r
barplot(table(ebola_data$district)) # bar plot of districts
```

```r
esquisse::esquisser(ebola_data) # interactive plotting with esquisse
```

```r
janitor::tabyl(ebola_data,sex)
```

```r
esquisse::esquisser(ebola_data)
```

```r
ggplot(ebola_data) +
  aes(x = sex, fill = sex) +
  geom_bar() +
  scale_fill_hue(direction = 1) +
  theme_minimal()
```

```r
ggplot(ebola_data) +
  aes(x = district, fill = district) +
  geom_bar() +
  scale_fill_hue(direction = 1) +
  theme_minimal()
```
