```r
if(!require(pacman) )
    install.packages("pacman")

pacman::p_load(tidyverse, visdat, esquisse, inspectdf, plotly, janitor)
```

```r
data <- read.csv("ebola_sierra_leone.csv")
View(data)
```

```r

head(data)
str(data)
```

```r
data$date_of_onset  <- as.Date(data$date_of_onset)
data$date_of_sample <- as.Date(data$date_of_sample)
```

```r
# 1 When was the first case reported?
first_case_date <- min(data$date_of_onset, na.rm = TRUE)

paste("The first case was reported on", first_case_date)
```

```r
# 2 What was the median age of those affected?

median_age <- median(data$age, na.rm = TRUE)

paste("The median age of those affected was", median_age)
```

```r
#3 Had there been more cases in men or women?

gender <- table(data$sex)
View(gender)
paste("There were more cases in", names(which.max(gender)))

```

```r
#4 What district had the most reported cases?

district_cases <- data %>%
    count(district) %>%
    arrange(desc(n))

View(district_cases)
```

```r
#5 By the end of June 2014, was the outbreak growing or receding?

cases_by_day <- data %>%
  count(date_of_onset)
june_cases <- cases_by_day %>%
  filter(date_of_onset <= as.Date("2014-06-30"))

ggplot(june_cases, aes(x = date_of_onset, y = n)) +
  geom_line() +
  labs(
    title = "Cases over time (up to end of June 2014)",
    x = "Date",
    y = "Number of cases"
  )
# The trend in cases up to the end of June 2014 appears to be growing, as indicated by the upward trajectory of the line plot.
```
